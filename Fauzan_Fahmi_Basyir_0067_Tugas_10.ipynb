Dictionary

```python

keranjang = {

    'apel' : 5000,
    'pisang' : 3000,
    'jeruk' : 7000,
    'anggur' : 15000

}

print('--------Kotok Mart--------')
total = 0
for buah, harga in keranjang.items():
  print(f'Harga {buah}\t: Rp.{harga}')
  total += harga
print('---------------------------')
print(f'Total Belanja\t: {total}')



```
